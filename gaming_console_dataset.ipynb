```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Columns:
- Console Name: The name of the game console.
- Type: The type of console (e.g., home console, handheld console).
- Company: The company that manufactured the console.
- Released Year: The year in which the console was first released.
- Discontinuation Year: The year in which the console was discontinued (if applicable).
- Units Sold (million): The total number of units sold worldwide, in millions.
- Remarks: Any additional remarks or comments about the console (if applicable).

```python
# Filter the dataframe to only include Nintendo consoles
def top_console(name): 
    nintendo_df = df[df['company'] == name]

# Group by released year and calculate total units sold for each year
    units_sold_by_year = nintendo_df.groupby('released_year')['units_sold_million'].sum()

    # Create a line plot to show the trend over time
    fig,axis = plt.subplots(figsize = (15,6))
    plt.plot(units_sold_by_year.index, units_sold_by_year.values, '-o')
    plt.plot(units_sold_by_year.index, units_sold_by_year.values, '-o', color='g')
    plt.xlabel('year')
    plt.ylabel('units_sold_million')
    plt.title('Nintendo Console Sales by Year')
    return plt.show()
```

```python
top_console('Sony')
```

```python
nintendo_df1 = df[df['company'] == 'Sony']
nintendo_df2 = df[df['company'] == 'Nintendo']

units_sold_by_year1 = nintendo_df1.groupby('released_year')['units_sold_million'].sum()
units_sold_by_year2 = nintendo_df2.groupby('released_year')['units_sold_million'].sum()

fig,axis = plt.subplots(figsize = (15,6))
plt.plot(units_sold_by_year1.index, units_sold_by_year1.values, '-o')
plt.plot(units_sold_by_year2.index, units_sold_by_year2.values, '-o', color='g')

plt.xlabel('year')
plt.ylabel('units_sold_million')
plt.legend(['Sony', 'Nintendo'], fontsize = 12)
plt.title('Console Sales by Year')
```
